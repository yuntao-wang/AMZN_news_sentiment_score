```python
import numpy as np
import pandas as pd
tweet_1 = pd.read_csv('amazon1.tweets.csv')
tweet_2 = pd.read_csv('amazon2.tweets.csv')
tweet_3 = pd.read_csv('amazon3.csv')
tweet_4 = pd.read_csv('amazon4.csv')
tweet_5 = pd.read_csv('amazon5.csv')
tweet_6 = pd.read_csv('amazon6.csv')
```

```python
tweet_1 = tweet_1[['created_at','text']]
tweet_2 = tweet_2[['created_at','text']]
tweet_3 = tweet_3[['created_at','text']]
tweet_4 = tweet_4[['created_at','text']]
tweet_5 = tweet_5[['created_at','text']]
tweet_6 = tweet_6[['created_at','text']]
```

```python
# combine tweets to get the tweets from 2018-03-03 to 2018-04-07
frames = [tweet_1, tweet_2,tweet_3,tweet_4,tweet_5, tweet_6]
tweets = pd.concat(frames)
tweets = tweets.reset_index()
del tweets['index']
time = tweets['created_at']
time_line=pd.to_datetime(time)
tweets['new_time']=time_line
tweets = tweets.sort_values(['new_time'],ascending=1)
```

```python
month_day = [('month'+str(x.month)+' '+ 'day' + str(x.day)) for x in tweets['new_time']]
tweets['month_day'] = month_day
```

```python
day_list = tweets['month_day'].unique().tolist()
day_list
```

```python
import nltk
from nltk.corpus import sentiwordnet as swn
def sentiment_score(doc):
    global doc_sentiment
    sentences = nltk.sent_tokenize(doc)
    stokens = [nltk.word_tokenize(sent) for sent in sentences]
    taggedlist=[]
    for stoken in stokens:        
         taggedlist.append(nltk.pos_tag(stoken))
    wnl = nltk.WordNetLemmatizer()

    score_list=[]
    for idx,taggedsent in enumerate(taggedlist):
        score_list.append([])
        for idx2,t in enumerate(taggedsent):
            newtag=''
            lemmatized=wnl.lemmatize(t[0])
            if t[1].startswith('NN'):
                newtag='n'
            elif t[1].startswith('JJ'):
                newtag='a'
            elif t[1].startswith('V'):
                newtag='v'
            elif t[1].startswith('R'):
                newtag='r'
            else:
                newtag=''       
            if(newtag!=''):    
                synsets = list(swn.senti_synsets(lemmatized, newtag))
                #Getting average of all possible sentiments, as you requested        
                score=0
                if(len(synsets)>0):
                    for syn in synsets:
                        score+=syn.pos_score()-syn.neg_score()
                    score_list[idx].append(score/len(synsets))
    sentence_sentiment=[]
    for score_sent in score_list:
        if len(score_sent)!=0:
            sentence_sentiment.append(sum([word_score for word_score in score_sent])/len(score_sent))
    if sentence_sentiment !=[]:
        doc_sentiment = sum(sentence_sentiment)/len(sentence_sentiment)
    return doc_sentiment
```

```python
tweet_score = []
for i in range(0,len(tweets)):   
    score = sentiment_score(tweets['text'][i])
    tweet_score.append(score)
```

```python
tweets['tweet_score'] = tweet_score
```

```python
score_day_list = []
for i in range(0,len(day_list)):
    score_list = []
    score_day = np.mean(tweets[tweets['month_day']==day_list[i]]['tweet_score'].tolist())
    score_day_list.append(score_day)
```
